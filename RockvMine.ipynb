Import Libraries and Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
sonar_data = pd.read_csv('/content/rockvmine.csv', header=None)
```

```python
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

So here we have counts of M (Mines) and R (Rocks)

```python
sonar_data.groupby(60).mean()
```

```python
features = sonar_data.drop(columns=60, axis=1)
label = sonar_data[60]
```

Splitting Feature Set and Label Set into Training (X_train and Y_train) and Test Set (X_test and Y_test) Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, label, test_size = 0.1, stratify=label, random_state=2)
```

Our Model is going to be Logistic Regression Model

```python
lrmodel = LogisticRegression()
```

Training our Model

```python
lrmodel.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_pred = lrmodel.predict(X_train)
train_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print(train_accuracy)
```

```python
X_test_pred = lrmodel.predict(X_test)
test_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print(test_accuracy)
```
